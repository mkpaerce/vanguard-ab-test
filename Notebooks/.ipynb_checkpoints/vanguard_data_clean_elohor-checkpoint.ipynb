```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df_clients = pd.read_csv("../Data/Raw/df_final_experiment_clients.txt")
df_demo = pd.read_csv("../Data/Raw/df_final_demo.txt")
df_web1 = pd.read_csv("../Data/Raw/df_final_web_data_pt_1.txt")
df_web2 = pd.read_csv("../Data/Raw/df_final_web_data_pt_2.txt")
```

```python
df_web=pd.concat([df_web1,df_web2], axis=0)
df_web=df_web.reset_index(drop=True)
df_web
```

```python
df_demo=pd.merge(df_demo,df_clients, on="client_id", how="left")
conditiontest=df_demo["Variation"]=="Test"
df_demo_test=df_demo[conditiontest]
df_demo_test
```

```python
conditioncontrol=df_demo["Variation"]=="Control"
df_demo_control=df_demo[conditioncontrol]
df_demo_control
```

```python
df_demo_test_clean = df_demo_test.dropna()
df_demo_test_clean
```

```python
#removing decimals/converting the columns from float to integers

df_demo_test_clean['clnt_age'] = df_demo_test_clean['clnt_age'].fillna(0).astype(int)
df_demo_test_clean['clnt_tenure_yr'] = df_demo_test_clean['clnt_tenure_yr'].fillna(0).astype(int)
df_demo_test_clean['clnt_tenure_mnth'] = df_demo_test_clean['clnt_tenure_mnth'].fillna(0).astype(int)
df_demo_test_clean['num_accts'] = df_demo_test_clean['num_accts'].fillna(0).astype(int)
df_demo_test_clean['calls_6_mnth'] = df_demo_test_clean['calls_6_mnth'].fillna(0).astype(int)
df_demo_test_clean['logons_6_mnth'] = df_demo_test_clean['logons_6_mnth'].fillna(0).astype(int)
df_demo_test_clean['bal'] = df_demo_test_clean['bal'].fillna(0).astype(int)
df_demo_test_clean
```

```python
#How long have the control clients had accounts?

disp_tenure_months = df_demo_test_clean['clnt_tenure_mnth'].describe()
disp_tenure_years = df_demo_test_clean['clnt_tenure_yr'].describe()
mode_tenure_years = df_demo_test_clean['clnt_tenure_yr'].mode()[0]
mode_tenure_months = df_demo_test_clean['clnt_tenure_mnth'].mode()[0]

disp_tenure_months, disp_tenure_years, mode_tenure_years, mode_tenure_months
```

```python
fig, axes = plt.subplots(2,1, dpi = 150)
sns.histplot(df_demo_test_clean['clnt_tenure_mnth'], kde=True, bins=30, color="purple", ax=axes[0])
axes[0].set_title("Histogram for Tenure Months, Test")
sns.histplot(df_demo_test_clean['clnt_tenure_yr'], kde=True, bins=30, color="pink", ax=axes[1])
axes[1].set_title("Histogram for Tenure Years, Test")
plt.tight_layout()
plt.show()
```
