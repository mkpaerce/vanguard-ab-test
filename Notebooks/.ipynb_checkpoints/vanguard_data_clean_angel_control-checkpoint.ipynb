## **Import the libraries** ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

## **Cleaning and columns aggregation** ##

```python
#Clean web_data_test 1148 duplicated values
df_web_control = pd.read_csv("../Data/Clean/df_final_web_data_control.csv")
df_web_control=df_web_control.reset_index(drop=True)
df_web_control.duplicated().sum()
df_web_control.drop_duplicates(inplace=True)
df_web_control.duplicated().sum()
```

```python
#Split date and time
df_web_control["date"]=df_web_control["date_time"].str.split(" ", expand=True)[0]
df_web_control["time"]=df_web_control["date_time"].str.split(" ", expand=True)[1]
df_web_control["hour"]=df_web_control["time"].str.split(":",expand=True)[0]
df_web_control["hour"]=df_web_control["hour"].astype(int)
df_web_control.head()
```

```python
#Change the dtypes of the column time %H,%M,%S format
df_web_control['time'] = (pd.to_datetime(df_web_control['time'], format='%H:%M:%S'))
df_web_control
```

```python
#Change the values of process_step
df_web_control["process_step"]=df_web_control["process_step"].map({"start":0,"step_1":1,"step_2":2,"step_3":3,"confirm":4})
df_web_control["process_step"].unique()
```

```python
#Create two new columns in order to have the step_jump and one column "control" to filter when the data is not from the same visit_id.
df_web_control=df_web_control.sort_values(["visit_id","date", "time", "process_step"])
df_web_control["step_jump"]=df_web_control["process_step"].astype(str).shift(+1)+"-"+df_web_control["process_step"].astype(str)
#Create column control which help me to determine, later, is the steps follows a orded sense.
df_web_control["control"]=(df_web_control["process_step"]-df_web_control["process_step"].shift(+1)).abs()
df_web_control["dif_time"]=(df_web_control["time"])-(df_web_control["time"].shift(+1))
df_web_control
```

```python
#Create a column for the total of segs spend between steps
df_web_control=df_web_control.dropna(subset=["control"])
df_web_control["dif_time"] = (df_web_control['dif_time'].astype(str))
df_web_control["dif_time"]=(df_web_control["dif_time"].str.split(" ")).str[-1]
df_web_control["dif_hour"]=((df_web_control["dif_time"].str.split(":")).str[0])
df_web_control["dif_hour"]=df_web_control["dif_hour"].astype(int)
df_web_control["dif_min"]=((df_web_control["dif_time"].str.split(":")).str[1]).astype(int)
df_web_control["dif_seg"]=((df_web_control["dif_time"].str.split(":")).str[2]).astype(int)
df_web_control["total_seg"]=df_web_control["dif_hour"]*3600 + df_web_control["dif_min"]*60 + df_web_control["dif_seg"]
df_web_control
```

```python
#Function for the keep_control column. It is only going to be used in terms of TIME BETWEEN STEPS
def row_count(df):
    df['next_visit'] = df['visit_id'].shift(+1) 
    df['keep_contol'] = df.apply(lambda row: "YES" if row['visit_id'] == row['next_visit'] else "NO", axis=1)
    df.drop(columns=['next_visit'], inplace=True)
    return df
```

```python
#Add the column keep_control to my DataFrame
df_web_control=row_count(df_web_control)
df_web_control

#FINAL DATA FRAME TO USE: df_web_control
```

## **Completion ratio** ##

```python
# Option 1: Based on total users to reached step 0
```

```python
counts_control = df_web_control['process_step'].value_counts()

counts_control
```

```python
#Completion ratio: number of visit in steo 0 / number of visit in step 4

from statsmodels.stats.proportion import proportions_ztest

x_test = 25721  # Number of completions in test group
n_test = 56474 # Total users in test group

x_control = 17505  # Number of completions in control group
n_control = 47607 # Total users in control group

completion_rate_test = x_test/n_test
completion_rate_control = x_control/n_control
completion_rate_test, completion_rate_control
```

```python
# Define alternative hypothesis ("larger" means one-tailed test)
count = np.array([x_test, x_control])  # Number of successes
nobs = np.array([n_test, n_control])  # Number of trials
z_stat, p_value = proportions_ztest(count, nobs, alternative="larger")

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The test group has a higher completion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in completion rates.")
```

## **Error Rate** ##

```python
#First we filter the values that we are going to use. Filter: Control=1; Keep=YES.
#We only want the errors in the process. 
condition_control=df_web_control["control"]==1
df_error_rate=df_web_control[condition_control]
condition_keep=df_error_rate["keep_contol"]=="YES"
df_error_rate=df_error_rate[condition_keep]
condition_step=(df_error_rate["step_jump"]=="1-0") | (df_error_rate["step_jump"]=="2-1") | (df_error_rate["step_jump"]=="3-2") | (df_error_rate["step_jump"]=="4-3")
df_error_rate=df_error_rate[condition_step]
```

```python
from statsmodels.stats.proportion import proportions_ztest

x_test = 11738  # Number of errors in test group df_error_rate.count()
n_test = 177846 # Total number of step jumps in test group based on df_web_test.count()

x_control = 6239  # Number of errors in control group df_error_rate.count()
n_control = 143461 # Total number of step jumps in control group based on df_web_control.count()

error_rate_test = x_test/n_test
error_rate_control = x_control/n_control
error_rate_test, error_rate_control
```

```python
# Define alternative hypothesis ("smaller" means one-tailed test)
count = np.array([x_test, x_control])  # Number of errors
nobs = np.array([n_test, n_control])  # Number of step jumps
z_stat, p_value = proportions_ztest(count, nobs, alternative="smaller")

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis: The test group has a lower error rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in error rates.")
```
